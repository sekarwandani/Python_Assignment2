```python
# menghapus kolom unnamed:0 dan ease-ment
df.drop(['Unnamed: 0','EASE-MENT'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# Mengubah tipe data pada kolom redidential units, commercial, total units, land square feet, gross square feet, dan sale price  menjadi data numeric
num = ["RESIDENTIAL UNITS","COMMERCIAL UNITS","TOTAL UNITS", "LAND SQUARE FEET" , "GROSS SQUARE FEET","SALE PRICE" ]

for col in num: 
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# mengubah kolom data sale date menjadi datatime
df['SALE DATE'] = pd.to_datetime(df['SALE DATE'], errors='coerce')
```

```python
# mengubah tipe data menjadi kategorikal
kateg = ['NEIGHBORHOOD', 'BUILDING CLASS CATEGORY', 'TAX CLASS AT PRESENT', 'BUILDING CLASS AT PRESENT', 'BUILDING CLASS AT TIME OF SALE', 'TAX CLASS AT TIME OF SALE']
for col in kateg:
    df[col] = df[col].astype('category')
```

```python
#cek missing value
missing = df.isnull().sum()
```

```python
#cek duplicated
df.duplicated().sum()
```

```python
# hapus data duplicated
df.drop_duplicates(inplace=True)

print(sum(df.duplicated()))
```

```python
#deskriptif
df[df['RESIDENTIAL UNITS']==0].describe().transpose()
```

```python
rep_mode = ['TAX CLASS AT PRESENT','BUILDING CLASS AT PRESENT']
for col1 in rep_mode:
    df[col1].fillna(df[col1].mode(), inplace=True)

#Tipe data numeric
df["LAND SQUARE FEET"]=df["LAND SQUARE FEET"].fillna(df["LAND SQUARE FEET"].median())
df["GROSS SQUARE FEET"]=df["GROSS SQUARE FEET"].fillna(df["GROSS SQUARE FEET"].median())
df["SALE PRICE"]=df["SALE PRICE"].fillna(df["SALE PRICE"].median())
```

```python
# menghapus data duplikat pada kolom YEAR BUILT
years = df["YEAR BUILT"].drop_duplicates()
years
```

```python
#Menghitung nilai 0 pada kolom YEAR BUILT, TOTAL UNIT, SALE PRICE
print("Nilai 0 pada YEAR BUILT: ",df[df['YEAR BUILT']==0]['YEAR BUILT'].count())
print("Nilai 0 pada TOTAL UNIT: ",df[df['TOTAL UNITS']==0]['TOTAL UNITS'].count())
print("Nilai 0 pada SALE PRICE: ",df[df['SALE PRICE']==0]['SALE PRICE'].count())
```

```python
#Menghapus nilai 0 pada kolom YEAR BUILT, TOTAL UNIT, SALE PRICE
df['YEAR BUILT'].replace(0, np.nan, inplace=True)
df['TOTAL UNITS'].replace(0, np.nan, inplace=True)
df['SALE PRICE'].replace(0, np.nan, inplace=True)
```

```python
#Pengecekan nilai 0 pada kolom YEAR BUILT, TOTAL UNIT, SALE PRICE 
print("Nilai 0 pada YEAR BUILT: ",df[df['YEAR BUILT']==0]['YEAR BUILT'].count())
print("Nilai 0 pada TOTAL UNIT: ",df[df['TOTAL UNITS']==0]['TOTAL UNITS'].count())
print("Nilai 0 pada SALE PRICE: ",df[df['SALE PRICE']==0]['SALE PRICE'].count())
```

```python
df = df[(df['SALE PRICE'] > 100000) & (df['SALE PRICE'] < 800000)]
```

```python
# Melakukan encode pada kolom BOROUGH
df['BOROUGH']= df['BOROUGH'].map({1:'Manhattan', 2:'Bronx', 3: 'Brooklyn', 4:'Queens',5:'Staten Island'})
df.head()
```

# Central Tendency

* Mean

```python
# menghitung Rata-rata setiap wilayah
rata_rata=df['SALE PRICE'].groupby(df['BOROUGH']).value_counts()
# Graded-Funtion Begin (~1 Lines)
print(rata_rata)
# Graded-Funtion End
print('mean',rata_rata.mean())
```

* Median

```python
# Menghitung Nilai tengah
nilai_tengah= df['SALE PRICE'].groupby(df['BOROUGH']).value_counts()

print(nilai_tengah.describe())

print('median',nilai_tengah.median())
```

* Modus

```python
# Menghitung modus

#modus = mode(data)
modus = df['SALE PRICE'].groupby(df['BOROUGH']).value_counts()
modus.mode()
```

* Range

```python
# Menghitung Range
Range = df['SALE PRICE'].groupby(df['BOROUGH']).value_counts()
Range=Range.describe()
range = Range['max']-Range['min']
range
```

* Variance

```python
# Menghitung Variance
varians = df['SALE PRICE'].groupby(df['BOROUGH']).value_counts()
varians.var()
```

* Standard Deviation

```python
standar_deviasi = df['SALE PRICE'].groupby(df['BOROUGH']).value_counts()

standar_deviasi.std()
```

```python
#distribusi pada kolom sale price
plt.figure(figsize=(20,15))
ax = sns.distplot(df['SALE PRICE'],
                  kde=True,
                  bins=100,
                  color='skyblue',
                  hist_kws={"linewidth":10, 'alpha':1})
ax.set(xlabel='sale price', ylabel='frequency')
plt.xlabel('Sale Price', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.title('Distribution of the Sale Price values', fontsize = 20)
plt.show()
```

```python
#distribusi tahun dibangun
sns.distplot(df['YEAR BUILT'], bins=20)
plt.show()
```

```python
# Distribution Probability of Sale Price in Bronx
plt.figure(figsize = (10,6))
distribusi_sp = df.groupby("BOROUGH")
n = distribusi_sp.get_group('Brooklyn')


ax = sns.distplot(n['SALE PRICE'],
                  kde=True,
                  bins=15,
                  color='orange',
                  hist_kws={"linewidth": 15,'alpha': 0.35})

ax.set(xlabel='Price', 
       ylabel='Probability of Occurence', 
       title='Distribution of Sale Price in Bronx')
```

```python
# Distribution of Sale Price in Brooklyn
plt.figure(figsize=(10,6))
df_1 = df.groupby("BOROUGH")
n = df_1.get_group('Brooklyn')


ax = sns.distplot(n['SALE PRICE'],
                  kde=True,
                  bins= 20,
                  color='purple',
                  hist_kws={"linewidth": 15,'alpha':0.55})

ax.set(xlabel='Price', 
       ylabel='Probability of Occurence', 
       title='Distribution of Sale Price in Brooklyn')
```

```python
#Mengambil log pada kolom sale price
plt.figure(figsize=(10,6))
sns.distplot(np.log(df['SALE PRICE'].values), kde=True)
plt.xlabel('Sale Price', fontsize = 20)
plt.ylabel('Frequency', fontsize = 20)
plt.xticks(fontsize = 12)
plt.yticks(fontsize = 12)
plt.title('Distribution of the Sale Price values', fontsize = 20)
plt.show()
```
